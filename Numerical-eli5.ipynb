```python
import pandas as pd
```

```python
df = pd.read_csv("movies_clean.csv")

df
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
df[['budget', 'popularity', 'revenue', 'runtime', 'vote_count', 'vote_average']].head()
```

```python
# Convert the target variable to a categorical variable
df['vote_average'] = pd.cut(df['vote_average'], bins=[0.,2.,4.,6.,8.,10.], labels=[0,1,2,3,4])
df = df.dropna(subset=['vote_average'])
y = df['vote_average']
# Select the required features
X = df[['budget', 'popularity', 'revenue', 'runtime', 'vote_count']]

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X.head()
```

```python
import numpy as np
np.unique(y)
```

```python
# Create an instance of the XGBoost Classifier model
model = xgb.XGBClassifier()

# Fit the model to the training data
model.fit(X_train, y_train)
```

```python
# Use the trained model to make predictions on the test set
y_pred = model.predict(X_test)

# Generate a classification report
report = classification_report(y_test, y_pred)

# Print the classification report
print(report)
```
